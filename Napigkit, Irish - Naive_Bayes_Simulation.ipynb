```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

# Dataset
data = pd.DataFrame({
    'Free': ['Yes', 'Yes', 'No', 'No', 'No'],
    'Win':  ['Yes', 'No', 'Yes', 'No', 'No'],
    'Class': ['Spam', 'Spam', 'Spam', 'Not Spam', 'Not Spam']
})

# Encode text to numbers
le = LabelEncoder()
data['Free'] = le.fit_transform(data['Free'])
data['Win'] = le.fit_transform(data['Win'])
data['Class'] = le.fit_transform(data['Class'])

# Features and target
X = data[['Free', 'Win']]
y = data['Class']

# Train model
model = GaussianNB()
model.fit(X, y)

# Predict new sample
prediction = model.predict([[1, 0]])

print("Prediction:", "Spam" if prediction[0] == 1 else "Not Spam")
```

## Naive Bayes Simulation

This simulation demonstrates the Naive Bayes algorithm for classifying emails as Spam or Not Spam.

The dataset includes two features:
- "Free" (whether the email contains the word free)
- "Win" (whether the email contains the word win)

The model was trained using Gaussian Naive Bayes and used to predict a new email.
The result shows whether the email is likely Spam or Not Spam.
